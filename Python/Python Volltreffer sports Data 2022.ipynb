```python
data['Year'] = data['Date'].df.year

# Group by year and sum the sales
yearly_sales = data.groupby('Year')['Amount'].sum()

print(yearly_sales)
```

```python
grouped_by_category = data.groupby('Category')['Amount'].sum()
grouped_by_category.sort_values(ascending=False)
```

```python

# Calculate total sales by month
total_sales_by_month = data.groupby('Month')['Amount'].sum().sort_values(ascending=False).reset_index()

# Create a line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Amount', data=total_sales_by_month, marker='o', color='red')
plt.title('Total Sales Trend by Month')
plt.xlabel('Month')
plt.ylabel('Total Amount in Million')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid for better visualization
plt.show()
```

#### The analysis reveals that March was the best-performing month in terms of sales, with a total revenue of 1,29,651.  This insight is crucial for resource allocation and strategic planning.

```python
total_sales_by_month = data.groupby('Month')['Amount'].sum().sort_values(ascending=False).reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Amount', data=total_sales_by_month,palette='viridis')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Amount in Million')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Group by 'City' and calculate the sum of 'Amount'
city_sales = data.groupby('Ship-State')['Amount'].sum().sort_values(ascending=False)

# Select the top 10 performing cities
top_10_cities = city_sales.head(10)

# Create a new DataFrame with the results
result_df = pd.DataFrame({
    'Ship-State': top_10_cities.index,
    'Amount': top_10_cities.values
})

# Visualize the results using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Ship-State', y='Amount', data=result_df, palette='viridis')
plt.title('Top 10 Performing State - Sales Analysis')
plt.xlabel('State')
plt.ylabel('Total Sales Amount in Million')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

#### Maharashtra emerged as the top-selling State, indicating a significant market presence. Understanding State-wise sales patterns can aid in targeted marketing and inventory management.

```python

# Group by 'City' and calculate the sum of 'Amount'
city_sales = data.groupby('Ship-City')['Amount'].sum().sort_values(ascending=False)

# Select the top 10 performing cities
top_10_cities = city_sales.head(10)

# Create a new DataFrame with the results
result_df = pd.DataFrame({
    'Ship-City': top_10_cities.index,
    'Amount': top_10_cities.values
})

# Visualize the results using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Ship-City', y='Amount', data=result_df)
plt.title('Top 10 Performing Cities - Sales Analysis')
plt.xlabel('City')
plt.ylabel('Total Sales Amount in Million')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
total_sales_by_category = data.groupby('Category')['Amount'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Amount', data=total_sales_by_category)
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Amount in Million')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
total_sales_by_gender = data.groupby('Gender')['Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))

# Plotting a pie chart
plt.pie(total_sales_by_gender['Amount'], labels=total_sales_by_gender['Gender'], autopct='%1.1f%%', startangle=90)

plt.title('Total Sales by Gender')
plt.show()
```

```python
total_sales_by_gender = data.groupby('Gender')['Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))

# Plotting a pie chart
plt.pie(total_sales_by_gender['Amount'], labels=total_sales_by_gender['Gender'], autopct='%1.1f%%', startangle=90)

plt.title('Total Sales by Gender')
plt.show()
```

```python
# Subset the data for women's and men's items
women_items = data[data['Gender'] == 'Women']
men_items = data[data['Gender'] == 'Men']

# Analyze the most popular women's items
women_popular_items = women_items.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(5)

# Analyze the most popular men's items
men_popular_items = men_items.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(5)


plt.figure(figsize=(14, 8))
# Women's items plot
plt.subplot(2, 2, 1)
sns.barplot(x=women_popular_items.index, y=women_popular_items.values, palette='viridis')
plt.title("Women's Most Popular Items")
plt.xlabel('Product Category')
plt.ylabel('Total Amount in Million')

# Men's items plot
plt.subplot(2, 2, 2)
sns.barplot(x=men_popular_items.index, y=men_popular_items.values, palette='viridis')
plt.title("Men's Most Popular Items")
plt.xlabel('Product Category')
plt.ylabel('Total Amount in Million')

plt.tight_layout()
plt.show()
```

```python
total_sales_by_size = data.groupby('Size')['Amount'].sum().sort_values(ascending=False).reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(x='Size', y='Amount', data=total_sales_by_size,)
plt.title('Total Sales by Size')
plt.xlabel('Size')
plt.ylabel('Total Amount in Million')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
total_sales_by_channel = data.groupby('Channel ')['Amount'].sum().reset_index()

plt.figure(figsize=(10, 6))

# Plotting a pie chart
plt.pie(total_sales_by_channel ['Amount'], labels=total_sales_by_channel ['Channel '], autopct='%1.1f%%', startangle=90)

plt.title('Total Sales by Channel')
plt.show()
```

```python
total_sales_by_status = data.groupby('Status')['Amount'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Status', y='Amount', data=total_sales_by_status,palette='viridis')
plt.title('Total Sales by Status')
plt.xlabel('Status')
plt.ylabel('Total Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
